```python
# using iloc for exclidung first column
df1 = df.iloc[:,1:]
```

# outlier detection

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outlier = ((df1 < lower_bound) | (df1 > upper_bound).sum())
print("outlier detected per column")
outlier
```

# cleaned outliers

```python
cleaned_outliers = df1[~((df1 < lower_bound) | (df1 > upper_bound)).any(axis=1)]
cleaned_outliers
```

```python
sns.boxplot(data=df1)
plt.show()
```

```python
sns.boxplot(data=cleaned_outliers)
plt.show()
```

```python
total_row = df1.shape[0]
total_row
```

```python
remaining_row = cleaned_outliers.shape[0]
remaining_row
```

```python
print(f"total row before outlier removal :{total_row}")
print(f"remaining row after outier removel: {remaining_row}")
print(f"percentage of data left {remaining_row / total_row * 100:.2f}%")
```

# feature target

```python
x = df1.iloc[:,:-1]
y = df1.iloc[:,-1]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

# pca for dim reduction

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
sns.scatterplot(x=x_pca[:,0],y=x_pca[:,1],hue=y,palette="coolwarm")
plt.xlabel("principal component 1")
plt.ylabel("principal componenet 2")
plt.title("PCA Visualization of Spectral Data")
```

# splitting data

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,train_size=0.2,random_state=42)
```

```python
model = RandomForestRegressor(random_state=42)
```

 # using hyper parameter tuning with RandomizedSearchCV

```python
param_grid = {
    "n_estimators":[100,200,300],
    "max_depth":[None,10,20],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
```

```python
model1 = RandomizedSearchCV(estimator=model,param_distributions=param_grid,n_iter=5,cv=3,verbose=2,n_jobs=-1)
model1.fit(x_train,y_train)
```

```python
# finding best params
print("best parameter found :")
model1.best_params_
```

# training final model with best param

```python
best_model = RandomForestRegressor(**model1.best_params_,random_state=42)
best_model.fit(x_train,y_train)
```

```python
# prediction
y_pred = best_model.predict(x_test)
```

# model evaluation

```python
print("MAE :",mean_absolute_error(y_test,y_pred))
```

```python
print("MSE :",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
print("R2_Score :",r2_score(y_test,y_pred))
```
